```python
from gensim.models import KeyedVectors
wv = KeyedVectors.load('vectors.kv')
```

```python
def sent_vec(sent):
    vector_size = wv.vector_size
    wv_res = np.zeros(vector_size)
    # print(wv_res)
    ctr = 1
    for w in sent:
        if w in wv:
            ctr += 1
            wv_res += wv[w]
    wv_res = wv_res/ctr
    return wv_res
```

```python
nlp = spacy.load('/home/riya/anaconda3/lib/python3.11/site-packages/en_core_web_sm/en_core_web_sm-3.7.1')
stop_words = nlp.Defaults.stop_words
print(stop_words)
punctuations = string.punctuation
print(punctuations)
```

```python
# Define a custom function to remove word before 1st space and excess white spaces
def process_string(input_string):
    # Find the index of the first space
    first_space_index = input_string.find(' ')
    
    # Remove the word before the first space
    if first_space_index != -1:
        input_string = input_string[first_space_index+1:]
    
    # Remove excess white spaces
    input_string = ' '.join(input_string.split())
    
    return input_string
```

```python
# Apply the custom function to each cell in the 'Column' column
sdg['Tentative Indicators'] = sdg['Tentative Indicators'].apply(process_string)
sdg['Targets'] = sdg['Targets'].apply(process_string) 
sdg
```

```python
# Creating our tokenizer function
def spacy_tokenizer(sentence):
    # Creating our token object, which is used to create documents with linguistic annotations.
    doc = nlp(sentence)



    #print(doc)
    #print(type(doc))

    # Lemmatizing each token and converting each token into lowercase
    mytokens = [ word.lemma_.lower().strip() for word in doc ]

    #print(mytokens)

    # Removing stop words
    mytokens = [ word for word in mytokens if word not in stop_words and word not in punctuations ]

    # return preprocessed list of tokens
    return mytokens
```

```python
# Function to remove duplicates from a list and return a list of unique words
def remove_duplicates(word_list):
    return list(set(word_list))
```

```python
sdg_new = sdg.groupby(['Goal No.', 'Goal', 'Targets',  'Target_id'])['Tentative Indicators'].apply(lambda x: ', '.join(x)).reset_index()
sdg_new
```

```python
sdg_new['concat_value'] = sdg_new['Targets'] + ' ' + sdg_new['Tentative Indicators']
sdg_new.drop(['Tentative Indicators'], axis=1,inplace=True)

sdg_new.rename(columns = {'concat_value':'Tentative Indicators'}, inplace = True) 
sdg_new
```

```python
sdg_new['new_targets'] = sdg_new['Tentative Indicators'].apply(spacy_tokenizer)
```

```python
# Apply the function to create the new column
sdg_new['tokens'] = sdg_new['new_targets'].apply(remove_duplicates)
```

```python
sdg_new['vec'] = sdg_new['tokens'].apply(sent_vec)
```

```python
# Extract vectors from DataFrame
vectors = np.array(sdg_new['vec'][:10,].to_list())
vec_len = np.size(vectors)
vec_len /= 300
print(vec_len)
# Compute cosine similarity between vectors
similarity_matrix = cosine_similarity(vectors, vectors)
print(similarity_matrix)
# Convert the similarity matrix to a DataFrame for visualization
similarity_df = pd.DataFrame(similarity_matrix, columns=range(int(vec_len)), index=range(int(vec_len)))
```
