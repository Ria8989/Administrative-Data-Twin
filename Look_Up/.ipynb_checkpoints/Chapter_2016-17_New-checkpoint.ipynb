```python
import gensim
from gensim.models import word2vec
from gensim.models.word2vec import Word2Vec
import matplotlib.pyplot as plt
import spacy
import string
from sklearn.metrics.pairwise import cosine_similarity
import pandas as pd
import numpy as np
```

```python
from gensim.models import KeyedVectors
wv = KeyedVectors.load('vectors.kv')
```

```python
def sent_vec(sent):
    vector_size = wv.vector_size
    wv_res = np.zeros(vector_size)
    # print(wv_res)
    ctr = 1
    for w in sent:
        if w in wv:
            ctr += 1
            wv_res += wv[w]
    wv_res = wv_res/ctr
    return wv_res
```

```python
nlp = spacy.load('/home/riya/anaconda3/lib/python3.11/site-packages/en_core_web_sm/en_core_web_sm-3.7.1')
stop_words = nlp.Defaults.stop_words
print(stop_words)
punctuations = string.punctuation
print(punctuations)
```

```python
# Creating our tokenizer function
def spacy_tokenizer(sentence):
    # Creating our token object, which is used to create documents with linguistic annotations.
    doc = nlp(sentence)



    #print(doc)
    #print(type(doc))

    # Lemmatizing each token and converting each token into lowercase
    mytokens = [ word.lemma_.lower().strip() for word in doc ]

    #print(mytokens)

    # Removing stop words
    mytokens = [ word for word in mytokens if word not in stop_words and word not in punctuations ]

    # return preprocessed list of tokens
    return mytokens
```

```python
# Function to remove duplicates from a list and return a list of unique words
def remove_duplicates(word_list):
    return list(set(word_list))
```

```python
sdg.drop(['Unnamed: 0'], axis=1,inplace=True)
sdg
```

```python
# Extract vectors from DataFrame
sdg_vectors = np.array(sdg['vec'].to_list())

print(len(sdg_vectors))
print(len(vectors))
# Compute cosine similarity between vectors
similarity_matrix_sdg = cosine_similarity(sdg_vectors,vectors)

vec_len = int(len(vectors))
sdg_len = int(len(sdg_vectors))

# Convert the similarity matrix to a DataFrame for visualization
similarity_df_sdg = pd.DataFrame(similarity_matrix_sdg, columns=range(vec_len), index=range(sdg_len))
```

```python
similarity_df_sdg
```

```python
def find_similarity_above_threshold(similarity_matrix, threshold, new_df, sdg):
    # Initialize empty lists to store chapter_id, chapter_name, and similarity values
    max_similarity_chapter_id = []
    max_similarity_chapter_name = []
    max_similarity_value = []
    
    # Iterate through the rows of the similarity matrix
    for row_idx, row in enumerate(similarity_matrix):
        # Find the indices where similarity values exceed the threshold
        above_threshold_indices = np.where(row > threshold)[0]
        
        # Check if there are any indices above the threshold
        if len(above_threshold_indices) > 0:
            # Get the index of the maximum similarity value above the threshold
            max_index = np.argmax(row)
            max_val = np.max(row)

            # Get the corresponding chapter_id and chapter_name from the new_df DataFrame
            chapter_id = new_df['Chapter_id'].loc[max_index]
            chapter_name = new_df['Chapter_name'].loc[max_index]

            # Append the chapter_id, chapter_name, and similarity value to the lists
            max_similarity_chapter_id.append(chapter_id)
            max_similarity_chapter_name.append(chapter_name)
            max_similarity_value.append(max_val)
        else:
            max_similarity_chapter_id.append(np.nan)
            max_similarity_chapter_name.append(np.nan)
            max_similarity_value.append(np.nan)
    
    # Add new columns to the sdg DataFrame
    sdg['Max_similarity_chapter_id'] = max_similarity_chapter_id
    sdg['Max_similarity_chapter_name'] = max_similarity_chapter_name
    sdg['Max_similarity_value'] = max_similarity_value

    return sdg
```

```python
# Example usage:
threshold_value = 0.4  # Define the threshold value

# Call the function to find chapter_ids with similarity greater than the threshold
sdg_new = find_similarity_above_threshold(similarity_matrix_sdg, threshold_value, new_df, sdg)
```

```python
sdg_new
```

```python
# Group by 'Goal No.' and find the maximum appearing Max_similarity_chapter_id
max_appearing_chapter_id = sdg_new.groupby('Goal No.')['Max_similarity_chapter_id'].apply(lambda x: x.mode()[0]).reset_index()
# Merge the DataFrame with the maximum appearing Max_similarity_chapter_id with the original DataFrame
sdg_new = sdg_new.merge(max_appearing_chapter_id, on='Goal No.', how='left', suffixes=('', '_max'))
sdg_new
```

```python
sdg_goal = sdg_new[['Goal No.', 'Goal', 'Target_id', 'Max_similarity_chapter_id', 'Max_similarity_value', 'Max_similarity_chapter_id_max']]
sdg_goal
```

```python
look_up = new_df[['Chapter_id', 'Chapter_name']]
```

```python
# Create a dictionary to store chapter details
chapter_details = {}

# Iterate over the rows of the SDG DataFrame
for index, row in sdg_goal.iterrows():
    chapter_id = row['Max_similarity_chapter_id_max']
    goal_no = row['Goal No.']
    
    if chapter_id not in chapter_details:
        chapter_details[chapter_id] = set([goal_no])  # Use set instead of list
    else:
        chapter_details[chapter_id].add(goal_no)  # Add to set

# Convert sets back to lists
chapter_details = {chapter_id: list(goals) for chapter_id, goals in chapter_details.items()}
```

```python
# Create a new column in the chapter DataFrame to contain the list of goal numbers
look_up['Goal_numbers'] = look_up['Chapter_id'].map(chapter_details)
```
