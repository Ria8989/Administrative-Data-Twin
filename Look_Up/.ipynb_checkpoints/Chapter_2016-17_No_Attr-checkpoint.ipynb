```python
import gensim
from gensim.models import word2vec
from gensim.models.word2vec import Word2Vec
import matplotlib.pyplot as plt
import spacy
import string
from sklearn.metrics.pairwise import cosine_similarity
import pandas as pd
import numpy as np
```

```python
from gensim.models import KeyedVectors
wv = KeyedVectors.load('vectors.kv')
```

```python
def sent_vec(sent):
    vector_size = wv.vector_size
    wv_res = np.zeros(vector_size)
    # print(wv_res)
    ctr = 1
    for w in sent:
        if w in wv:
            ctr += 1
            wv_res += wv[w]
    wv_res = wv_res/ctr
    return wv_res
```

```python
nlp = spacy.load('/home/riya/anaconda3/lib/python3.11/site-packages/en_core_web_sm/en_core_web_sm-3.7.1')
stop_words = nlp.Defaults.stop_words
print(stop_words)
punctuations = string.punctuation
print(punctuations)
```

```python
# Creating our tokenizer function
def spacy_tokenizer(sentence):
    # Creating our token object, which is used to create documents with linguistic annotations.
    doc = nlp(sentence)



    #print(doc)
    #print(type(doc))

    # Lemmatizing each token and converting each token into lowercase
    mytokens = [ word.lemma_.lower().strip() for word in doc ]

    #print(mytokens)

    # Removing stop words
    mytokens = [ word for word in mytokens if word not in stop_words and word not in punctuations ]

    # return preprocessed list of tokens
    return mytokens
```

```python
# Function to remove duplicates from a list and return a list of unique words
def remove_duplicates(word_list):
    return list(set(word_list))
```

```python
new_df = data.groupby(['Chapter_id', 'Chapter_name'])['Table_name'].apply(lambda x: ', '.join(x)).reset_index()
new_df
```

```python
new_df.columns = ['Chapter_id', 'Chapter_name', 'Description']
```

```python
new_df['new_description'] = new_df['Description'].apply(spacy_tokenizer)
```

```python
# Apply the function to create the new column
new_df['tokens'] = new_df['new_description'].apply(remove_duplicates)
```

```python
new_df['vec'] = new_df['tokens'].apply(sent_vec)
```

```python
# Extract vectors from DataFrame
vectors = np.array(new_df['vec'].to_list())

# Compute cosine similarity between vectors
similarity_matrix = cosine_similarity(vectors, vectors)

# Convert the similarity matrix to a DataFrame for visualization
similarity_df = pd.DataFrame(similarity_matrix, columns=range(18), index=range(18))
```

```python
sdg = pd.read_csv('sdg_data_vec.csv', sep=";")
sdg
```

```python
# Convert string representations of vectors to actual NumPy arrays
sdg['vec'] = sdg['vec'].apply(lambda x: np.array([float(value) for value in x.strip('[]').split()]))
sdg
```

```python
# Extract vectors from DataFrame
sdg_vectors = np.array(sdg['vec'].to_list())

# Compute cosine similarity between vectors
similarity_matrix_sdg = cosine_similarity(vectors, sdg_vectors)

# Convert the similarity matrix to a DataFrame for visualization
similarity_df_sdg = pd.DataFrame(similarity_matrix_sdg, columns=range(18), index=range(18))
```

```python
similarity_df_sdg
```

```python
# Initialize empty lists to store chapter_id and chapter_name
max_similarity_chapter_id = []
max_similarity_chapter_name = []

# Iterate through the rows of the similarity matrix
for row in similarity_matrix_sdg:
    # Find the index of the maximum similarity value in the row
    max_index = np.argmax(row)
    
    # Get the corresponding chapter_id and chapter_name from the sdg DataFrame
    chapter_id = new_df['Chapter_id'].loc[max_index]
    chapter_name = new_df['Chapter_name'].loc[max_index]
    
    # Append the chapter_id and chapter_name to the lists
    max_similarity_chapter_id.append(chapter_id)
    max_similarity_chapter_name.append(chapter_name)

# Add new columns to the sdg DataFrame
sdg['Max_similarity_chapter_id'] = max_similarity_chapter_id
sdg['Max_similarity_chapter_name'] = max_similarity_chapter_name
```
