```python
import re

def extract_chapter_number(file_path):
    # Define a regular expression pattern to match the chapter number
    pattern = re.compile(r'_Chapter(\d+).pdf')

    # Search for the pattern in the file path
    match = pattern.search(file_path)

    if match:
        # Extract the chapter number from the matched group
        chapter_number = match.group(1)
        return chapter_number
    else:
        # Return None if the pattern is not found
        return None
```

```python
input_folder = '/kaggle/input/data-chap9-18/'

# Iterate through each file in the input folder
for root, dirs, files in os.walk(input_folder):
    !rm -rf /kaggle/working/*
    for filename in files:
        pdf_path = os.path.join(root, filename)
        print("Processing file:", pdf_path)
        chapter_number = extract_chapter_number(pdf_path)
        !rm -rf /kaggle/working/out
        os.mkdir(r'/kaggle/working/out')
        os.mkdir(r'/kaggle/working/out/pages')
        os.mkdir(r'/kaggle/working/out/output')
        doc = aw.Document()
        # Use PyPDF2 to get the number of pages in the PDF
        with open(pdf_path, "rb") as pdf_file:
            pdf_reader = PdfReader(pdf_file)
            num_pages = len(pdf_reader.pages)

        # Loop through each page and save it as a separate HTML file.
        for page_index in range(num_pages):
            # Load the PDF document for each page
            with open(pdf_path, "rb") as pdf_file:
                pdf_reader = PdfReader(pdf_file)
                pdf_writer = PdfWriter()
                pdf_writer.add_page(pdf_reader.pages[page_index])

                # Save the PDF for the current page
                page_pdf_path = f"/kaggle/working/out/pages/page_{page_index + 1}.pdf"
                with open(page_pdf_path, "wb") as page_pdf_file:
                    pdf_writer.write(page_pdf_file)

            # Load the page PDF with Aspose.Words
            doc = aw.Document(page_pdf_path)

            # Save the current page to HTML format.
            output_file_path = f"/kaggle/working/out/output/output_page_{page_index + 1}.html"
            doc.save(output_file_path, aw.SaveFormat.HTML)

            print(f"Page {page_index + 1} saved as {output_file_path}")
        
        !rm -rf /kaggle/working/out/output/*.png
        name = "Output_html_Chapter" + str(chapter_number)
        !zip -r {name}.zip /kaggle/working/out
```
