# Merging table header

```python
import pandas as pd
import os
```

```python
path = "/home/riya/Downloads/ADT/Dictionary/"
```

```python
os.chdir(path)
```

```python
dfs = []

# Iterate over each file in the directory
for filename in os.listdir():
    if filename.endswith('.csv'):
        file_path = os.path.join(path, filename)
        # Read the CSV file into a DataFrame without header
        df = pd.read_csv(file_path, sep=';')
        dfs.append(df)

# Concatenate all DataFrames into a single DataFrame
merged_df = pd.concat(dfs, ignore_index=True)

# Display the merged DataFrame
merged_df.columns = ['Attr_id', 'Chapter_id', 'Chapter_name', 'Table_id', 'Table_name', 'Description']
merged_df
```

```python
merged_df = merged_df.sort_values(by=['Attr_id'])
merged_df.to_csv('/home/riya/Downloads/ADT/Dictionary/Attributes.csv', sep=';', index=False)
```
