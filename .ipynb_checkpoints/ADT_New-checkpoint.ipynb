```python
import PyPDF2
import tabula

def extract_text_from_pdf(pdf_path):
    # Extract text from the PDF using PyPDF2
    with open(pdf_path, 'rb') as file:
        pdf_reader = PyPDF2.PdfReader(file)
        text = ''
        for page_num in range(len(pdf_reader.pages)):
            page = pdf_reader.pages[page_num]
            text += page.extract_text()
        return text

def extract_tabular_data_from_pdf(pdf_path):
    # Extract tabular data using tabula-py
    tables = tabula.read_pdf(pdf_path, pages='all', multiple_tables=True)
    return tables
```

```python
# Extract text from the PDF
pdf_text = extract_text_from_pdf(pdf_path)
print("Text Extracted from PDF:")
print(pdf_text)
```

```python
# Extract tabular data from the PDF
pdf_tables = extract_tabular_data_from_pdf(pdf_path)
print("\nTabular Data Extracted from PDF:")
print(pdf_tables)
```

```python
import pdfplumber

def extract_table_from_pdf(pdf_path):
    with pdfplumber.open(pdf_path) as pdf:
        # Assuming the table is on the first page, adjust page number as needed
        page = pdf.pages[0]
        table = page.extract_table()

        # Iterate through the rows of the table
        for row in table[1:]:  # Skip the header row
            # Print each cell value in the row
            print("|".join(str(cell) for cell in row))

# Extract and print the table
extract_table_from_pdf(pdf_path)
```
