# Merging table header

```python
import pandas as pd
import os
```

```python
relative_path_data = "Data/"
relatice_path_result = "Data/Data_2019-20.csv"
absolute_path_data = os.path.join(os.getcwd(), relative_path_data)
absolute_path_result = os.path.join(os.getcwd(), relatice_path_result)
```

```python
os.getcwd()
```

```python
absolute_path_data
```

```python
def concat_dataframe(df_list):
    df = pd.DataFrame()
    if len(df_list)>0:
        df = pd.concat([df, df_list[0].iloc[:,:2]], axis = 1)
        for item in df_list:
            df = pd.concat([df, item.iloc[:,2:]], axis = 1)
    return df
```

```python
dfs = []

# Iterate over each file in the directory
for filename in os.listdir(absolute_path_data):
    if filename.endswith('.csv'):
        file_path = os.path.join(absolute_path_data, filename)
        # Read the CSV file into a DataFrame without header
        df = pd.read_csv(file_path, sep=';')
        dfs.append(df)

# Concatenate all DataFrames into a single DataFrame
merged_df = concat_dataframe(dfs)

# Display the merged DataFrame
merged_df
```

```python
merged_df.shape
```

```python
final_df = merged_df.reindex(sorted(merged_df.columns, key=int), axis=1)
```

```python
final_df
```

```python
final_df.to_csv(absolute_path_result, sep=';', index=False)
```
