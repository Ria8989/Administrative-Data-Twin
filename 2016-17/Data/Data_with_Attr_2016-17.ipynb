```python
def concatenate_all(df):
    value = []
    for index, row in df.iterrows():
        value.append(str(row['Attr_id']) + ', ' + row['Description'] + ', ' + row['Table_name'] + ', ' + row['Chapter_name'])
    return value
```

```python
header = concatenate_all(lf)
lf['new_desc'] = header
```
