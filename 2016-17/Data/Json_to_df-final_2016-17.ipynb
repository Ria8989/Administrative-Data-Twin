```python
#Updating 1st disordered sequence

def interchange_rows(df_list, idx_list, n_order1):
    new_df_list = []
    for idx in range(len(df_list)):
        df = df_list[idx]
        if idx in idx_list:
            df.iloc[:,0] = n_order1
            sorted_df = df.sort_values(by=['1'], ascending=True)
            new_df_list.append(sorted_df.reset_index(drop=True))
        else:
            new_df_list.append(df)
    return new_df_list
```

```python
ordered_list = interchange_rows(new_list, index_list1, new_order1)
```

```python
final_list = interchange_rows(ordered_list, index_list2, new_order2)
```
