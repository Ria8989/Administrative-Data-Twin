```python
def interchange_rows(df, original_order, new_order, columns_list):
    # Check if the lengths of original_order and new_order are consistent
    if len(original_order) != len(new_order):
        raise ValueError("Lengths of original_order and new_order must be the same.")
    
    # Iterate over the columns in columns_list
    for col_name in columns_list:
        # Extract the original and new orders for the current column
        original_order_col = df.loc[original_order, col_name].copy()
        new_order_col = original_order_col.iloc[new_order].values
        # Update the dataframe with the new order
        df.loc[original_order, col_name] = new_order_col
    
    return df
```

```python
ordered_df = interchange_rows(df_org, original_order, new_order1, columns_list1)
```
